```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
```

```python
# creating 50 random points

# ref: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html
x, y = make_blobs(n_samples=50, centers=2, random_state=20)
# ref: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
classifier = svm.SVC(kernel='linear', C=1)
classifier.fit(x, y)
```

```python
# display the data

# ref: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
plt.scatter(x[:,0], x[:,1], c=y, cmap=plt.cm.Paired)
plt.show()
```
